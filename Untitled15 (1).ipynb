```python
def get_the_data(path='/content/usgs_data_big_file2 (1).csv'):
  l=[]
  with open(path,encoding='CP949') as fileob:
      fileob.readline()
      for line in fileob:
          l.append(dict(list(zip(columns,line.split(',')))))
  return pd.DataFrame(l,columns=columns)
```
